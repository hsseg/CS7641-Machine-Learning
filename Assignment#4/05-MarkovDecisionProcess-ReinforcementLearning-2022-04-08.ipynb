## 4x4 frozen lake environment

Imagine, there is a frozen lake from your home to office, you should walk on the frozen lake to reach your office. But oops! there will be a hole in the frozen lake in between, so you have to be careful while walking in the frozen lake to avoid getting trapped at holes.

- SFFF (S: starting point, safe)
- FHFH (F: frozen surface, safe)
- FFFH (H: hole, fall to your doom)
- HFFG (G: goal, where the frisbee is located)

Okay, now let us use our agent instead of you to find the correct way to reach the office. The agent goal is to find the optimal path to reach from S to G without getting trapped at H. How an agent can achieve this? We give +1 point as a reward to the agent if it correctly walks on the frozen lake and 0 points if it falls into the hole. So that agent could determine which is the right action. An agent will now try to find the optimal policy. Optimal policy implies taking the correct path which maximizes the agent reward. If the agent is maximizing the reward, apparently agent is learning to skip the hole and reach the destination.

Up, Down, Left, and Right directions are numerically represented below
- LEFT = 0
- DOWN = 1
- RIGHT = 2
- UP = 3

```python
# initialize the FrozenLake environment
env = gym.make('FrozenLake8x8-v0')
print(env.action_space)
print(env.observation_space)
print(env.reward_range)
```

```python
from lrl import environments, solvers
from lrl.utils import plotting
```

```python
lake = environments.frozen_lake.RewardingFrozenLakeEnv(map_name='4x4', is_slippery=False)

plotting.plot_env(env=lake)
```

#### Value Iteration (VI)

```python
lake_vi = solvers.ValueIteration(env=lake)
lake_vi.iterate_to_convergence()
```

```python
# max_iters = 500
scores = lake_vi.score_policy(iters=500)
scores_df = scores.to_dataframe(include_episodes=True)

print(scores_df.shape)
scores_df.head(3)
```

```python
scores_df.tail(3)
```

```python
ax_episodes = plotting.plot_episodes(episodes=scores.episodes, env=lake, max_episodes=100, )
```

```python
ax_results = plotting.plot_solver_results(env=lake, solver=lake_vi)
```

#### Policy Iteration (PI)

```python
lake_pi = solvers.PolicyIteration(env=lake)
lake_pi.iterate_to_convergence()
```

```python
# max_iters = 500
scores_pi = lake_pi.score_policy(iters=500)
scores_Pi_df = scores_pi.to_dataframe(include_episodes=True)

print(scores_Pi_df.shape)
scores_Pi_df.head(3)
```

```python
ax_results = plotting.plot_solver_results(env=lake, solver=lake_pi)
```

#### Compare max value change per iteration

```python
'''
>>> Compare the maximum change in value function
>>> Compare iterations to converge

'''
# ploting
plt.style.use('fivethirtyeight')

# Value iteration (vi) -- max value change
ax = plotting.plot_solver_convergence(lake_vi, label='vi')

# Policy iteration (pi) -- max value change
ax = plotting.plot_solver_convergence_from_df(lake_pi.iteration_data.to_dataframe(), y='delta_max', x='iteration', ax=ax, label='pi', ls='--', marker='o', markersize=10)

ax.legend()
```

#### Compare time per iteration

```python
'''
>>> Compare number of policy changes per iteration

'''

# Value iteration (vi) -- policy changes
ax = plotting.plot_solver_convergence(lake_vi, y='time', label='vi')

# Policy iteration (pi) -- policy changes
ax = plotting.plot_solver_convergence_from_df(lake_pi.iteration_data.to_dataframe(), y='time', x='iteration', ax=ax, label='pi', ls='--', marker='o', markersize=10)

ax.legend()
```

#### Q-Learning

```python
# Set hyper-parameters
#------------------------------------------------------
# Constant alpha during learning
alpha = 0.1

# define a linear decay function for epsilon
epsilon = {
  'type': 'linear',
  'initial_value': 0.2,
  'initial_timestep': 0,
  'final_value': 0.05,
  'final_timestep': 1500
  
}

# use default gamma value
gamma = 0.9

# parameters to monitor convergence
# Thus periodically evaluate how good the greedy policy at a given point in the solution
score_while_training = {
  'n_trains_per_eval': 50, # training episodes prior to scoring the greedy policy
  'n_evals': 250, # runs to score the greedy policy
}
```

```python
# define the Q-Learner for FrozenLake
lake_ql = solvers.QLearning(env=lake,
                           alpha=alpha,
                           epsilon=epsilon,
                           gamma=gamma,
                           min_iters=500,
                           max_iters=5000,
                           score_while_training=score_while_training)

lake_ql.iterate_to_convergence()
```

```python
ax_results = plotting.plot_solver_results(env=lake, solver=lake_ql)
```

```python
lake_ql_iter_df = lake_ql.iteration_data.to_dataframe()

print(lake_ql_iter_df.shape)
lake_ql_iter_df.head(10)
```

```python
# ploting
plt.style.use('fivethirtyeight')

lake_ql_iter_df.plot(x='iteration', y='policy_changes', kind='scatter', )
```

```python
lake_ql_iter_df.plot(x='iteration', y='delta_max', kind='scatter', )
```

```python
lake_ql_iter_df.plot(x='iteration', y='time', kind='scatter', )
```

```python
lake_ql.num_episodes_for_convergence
```
